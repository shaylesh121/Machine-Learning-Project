**Libraries that can be used for image processing**

1. matplotlib.image
2. Pillow
3. OpenCV (cv2)

```python
# importing the image module from matplotlib library

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# loading an image through matplotlib.image module

img = mpimg.imread('white-maltese-puppy-posing-photo-260nw-2481313223.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
print(img)
```

```python
# displaying the image from numpy array

img_plot = plt.imshow(img)
plt.show()
```

Resizing the image using Pillow library

```python
from PIL import Image as PILImage
```

```python
img = PILImage.open('white-maltese-puppy-posing-photo-260nw-2481313223.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('dog_img_resized.jpg')
```

```python
img = mpimg.imread('dog_img_resized.jpg')
img_plot = plt.imshow(img)
plt.show()
```

```python
img.shape
```

**Converting RGB images to Grayscale image using OpenCV**

```python
# importing OpenCV library
import cv2
```

```python
img = cv2.imread('white-maltese-puppy-posing-photo-260nw-2481313223.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image. But this will not be allowed in Jupyter notebook.

from google.colab.patches import cv2_imshow

but in jupyter notebook we need to use `IPython.display`

```python
# displaying the gray scale image
from IPython.display import display, Image
display(PILImage.fromarray(grayscale_image))
```

```python
# saving the grayscale image
cv2.imwrite('dog_grayscale_image.jpg', grayscale_image)
```
